```python
page_no = 1
page_url = f"https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page={page_no}"
page_url
```

```python
# !pip install requests 다운로드 받는법
```

```python
import requests

source = requests.get(page_url).text
source #400 404 405는 url이 유효하지 않는 경우이다.
```

```python
import bs4
source = bs4.BeautifulSoup(source)
source
```

```python
print(source.prettify())
```

```python
source.find_all('td', class_='number_1') #class 속성에 _를 붙이는 건 특별한 문법이다.
```
