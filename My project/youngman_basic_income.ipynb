```python
def youngman_basic_income(pIndex, pSize):
    KEY = 'f4b917c9fa67447d96c3114f3f002582'
    Type = 'json'
    url = f"https://openapi.gg.go.kr/YoungmanBasicIncome?KEY={KEY}&Type={Type}&pIndex={pIndex}&pSize={pSize}"
    response = requests.get(url)
    return response.json()
```
